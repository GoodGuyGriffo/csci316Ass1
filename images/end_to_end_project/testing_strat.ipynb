```python
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from sklearn.datasets import make_regression, make_classification
import numpy as np

seed = 23
n_samples, n_features = 10000, 10
rng = np.random.RandomState(seed)
iris = datasets.load_iris()
```

```python
iris_ = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
iris_["target"] = iris.target
```

```python
rnd_train_iris, rnd_test_iris = train_test_split(iris_, test_size=1/3, random_state=seed)
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=1/3, random_state=seed)
```

```python
for train_index, test_index in split.split(iris_, iris_.target):
    strat_train_iris = iris_.loc[train_index]
    strat_test_iris = iris_.loc[test_index]
```
