## Comparing the ratio from training set to the entire set

```python
def income_cat_proportions(data_loc):
    return data_loc["critical_temp_cat"].value_counts() / len(data_loc)
```

```python
train_set, test_set = train_test_split(df, test_size=0.33, random_state=seed)
```

```python
compare_props = pd.DataFrame({
    "Overall": income_cat_proportions(df),
    "Stratified": income_cat_proportions(strat_test_set),
    "Random": income_cat_proportions(test_set),
}).sort_index()
compare_props["Rand. %error"] = 100 * compare_props["Random"] / compare_props["Overall"] - 100
compare_props["Strat. %error"] = 100 * compare_props["Stratified"] / compare_props["Overall"] - 100
```

# Prepare the Data for Machine Learning Algoriuthms

```python
data = strat_train_set.drop("critical_temp", axis=1)
data = data.drop("critical_temp_cat", axis=1)
data_labels = strat_train_set["critical_temp"].copy()
```

```python
data
```

## DATA CLEANING

```python
sample_incomplete_rows = data[data.isnull().any(axis=1)].head()
sample_incomplete_rows
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler()),
    ])

data_prepared = pipeline.fit_transform(data)
```

```python
data_prepared
```

# Select and Train a Model

## Training linear regression

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(data_prepared, data_labels)
```

```python
lin_data_predictions = lin_reg.predict(data_prepared)
```

```python
from sklearn.metrics import mean_absolute_error

lin_mae = mean_absolute_error(data_labels, lin_data_predictions)
lin_mae
```

```python
from sklearn.metrics import mean_squared_error

lin_mse = mean_squared_error(data_labels, lin_data_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

## Training Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(random_state=seed)
tree_reg.fit(data_prepared, data_labels)
```

```python
tree_data_predictions = tree_reg.predict(data_prepared)
```

```python
tree_mse = mean_squared_error(data_labels, tree_data_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

## Training Random Forrest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(n_estimators=100, random_state=seed)
forest_reg.fit(data_prepared, data_labels)
```

```python
forest_data_predictions = forest_reg.predict(data_prepared)
```

```python
forest_mse = mean_squared_error(data_labels, forest_data_predictions)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

## Training Ridge model

```python
from sklearn import linear_model

ridge_reg = linear_model.Ridge(alpha=.5, random_state=seed)
ridge_reg.fit(data_prepared, data_labels)
```

```python
ridge_data_predictions = ridge_reg.predict(data_prepared)
```

```python
ridge_mse = mean_squared_error(data_labels, ridge_data_predictions)
ridge_rmse = np.sqrt(ridge_mse)
ridge_rmse
```

## Lasso Model

```python
lasso_reg = linear_model.Lasso(alpha=0.1, random_state=seed)
lasso_reg.fit(data_prepared, data_labels)
```

```python
lasso_data_predictions = lasso_reg.predict(data_prepared)
```

```python
lasso_mse = mean_squared_error(data_labels, lasso_data_predictions)
lasso_rsme = np.sqrt(lasso_mse)
lasso_rsme
```

## Cross Validation time

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
```

```python
from sklearn.model_selection import cross_val_score
```

```python
lin_scores = cross_val_score(lin_reg, data_prepared, data_labels,
                             scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

```python
tree_scores = cross_val_score(tree_reg, data_prepared, data_labels,
                         scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-tree_scores)
display_scores(tree_rmse_scores)
```

```python
forest_scores = cross_val_score(forest_reg, data_prepared, data_labels,
                         scoring="neg_mean_squared_error", cv=5)
forest_rmse_scores = np.sqrt(-forest_scores)
display_scores(forest_rmse_scores)
```

```python
ridge_scores = cross_val_score(ridge_reg, data_prepared, data_labels,
                             scoring="neg_mean_squared_error", cv=10)
ridge_rmse_scores = np.sqrt(-ridge_scores)
display_scores(lin_rmse_scores)
```

```python
lasso_scores = cross_val_score(lasso_reg, data_prepared, data_labels,
                             scoring="neg_mean_squared_error", cv=10)
lasso_rmse_scores = np.sqrt(-lasso_scores)
display_scores(lasso_rmse_scores)
```
